```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
# Veri setini yükleme
file_path = 'musicdataset.csv'
data = pd.read_csv(file_path)
```

```python
# 1. Sütunları temizleme: Gerekli olmayan sütunların çıkarılması
columns_to_drop = ['id']  # Benzersiz kimlik kullanmayacağız
data = data.drop(columns=columns_to_drop)
```

```python
# 2. Tarih formatını düzenleme
# Eğer release_date tam bir tarih formatında değilse sadece yıl olarak bırakabiliriz.
data['release_date'] = pd.to_datetime(data['release_date'], errors='coerce')
data['release_year'] = data['release_date'].dt.year
```

```python
# 3. Sanatçıların ayrıştırılması
# artists sütunundaki verileri temizleyip listelerden ilk sanatçıyı alabiliriz
data['artists'] = data['artists'].str.strip("[]").str.replace("'", "")
data['artists'] = data['artists'].str.split(',').str[0]
```

```python
missing_values = data.isnull().sum()
```

```python
print("Eksik Değerler:\n", missing_values)
```

```python
# 5. Özelliklerin ölçeklenmesi
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
numerical_features = [
    'valence', 'acousticness', 'danceability', 'duration_ms', 'energy',
    'instrumentalness', 'liveness', 'loudness', 'speechiness', 'tempo'
]
data[numerical_features] = scaler.fit_transform(data[numerical_features])
```

```python
# 6. Temizlenmiş veri setini kaydetme
cleaned_file_path = 'Cleaned_Music_Dataset.csv'
data.to_csv(cleaned_file_path, index=False)
print('Temizlenmiş veri seti kaydedildi:', cleaned_file_path)
print("Temizlenmiş veri setinin ilk 5 satırı:\n", data.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Korelasyon matrisini hesapla
correlation_matrix = data[numerical_features].corr()

# Matrisin görselleştirilmesi için bir ısı haritası
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', cbar=True)
plt.title("Korelasyon Matrisi")
plt.show()

```

```python
# Korelasyon matrisi hesaplandıktan sonra
correlation_threshold = 0.2  # İstenilen korelasyon eşiği
low_correlation_features = correlation_matrix.columns[correlation_matrix.abs().max(axis=0) < correlation_threshold]

# Bu özellikleri veri setinden çıkaralım
data_cleaned_reduced = data.drop(columns=low_correlation_features)

# Yeni veri setini kaydedelim
cleaned_reduced_file_path = 'Cleaned_Music_Dataset_Reduced.csv'
data_cleaned_reduced.to_csv(cleaned_reduced_file_path, index=False)
print('Düşük korelasyona sahip özellikler çıkarıldı ve yeni veri seti kaydedildi:', cleaned_reduced_file_path)

# Temizlenmiş verinin ilk 5 satırını yazdırma
print("Yeni veri setinin ilk 5 satırı:\n", data_cleaned_reduced.head())
```

```python

# Sayısal özelliklerin tekrar tanımlanması
numerical_features = [
    'valence', 'acousticness', 'danceability', 'duration_ms', 'energy',
    'instrumentalness', 'liveness', 'loudness', 'speechiness', 'tempo'
]
```

```python
# Popülerlik tahmini için özellik ve hedef ayrımı
X = data[numerical_features]
y = data['popularity']
```

```python
# Düşük korelasyona sahip özellikleri çıkardıktan sonra
X_cleaned = data_cleaned_reduced[numerical_features]  # yeni özelliklerle eğitim verisi
y = data['popularity']
```

```python
# Veriyi eğitim ve test setlerine ayırma (yeni özellik setini kullanarak)
X_train, X_test, y_train, y_test = train_test_split(X_cleaned, y, test_size=0.2, random_state=42)
```

```python
# 1. Ağaç sayısını azaltma (Random Forest)
rf_model_reduced = RandomForestRegressor(n_estimators=10, random_state=42)
rf_model_reduced.fit(X_train, y_train)
rf_pred_reduced = rf_model_reduced.predict(X_test)
rf_mse_reduced = mean_squared_error(y_test, rf_pred_reduced)
rf_rmse_reduced = rf_mse_reduced ** 0.5
```

```python
# 2. Veri alt kümesi kullanma
X_train_small = X_train.sample(frac=0.1, random_state=42)
y_train_small = y_train[X_train_small.index]

rf_model_small = RandomForestRegressor(random_state=42)
rf_model_small.fit(X_train_small, y_train_small)
rf_pred_small = rf_model_small.predict(X_test)
rf_mse_small = mean_squared_error(y_test, rf_pred_small)
rf_rmse_small = rf_mse_small ** 0.5
```

```python
# 3. Daha hızlı bir model kullanma (Linear Regression)
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)
lr_mse = mean_squared_error(y_test, lr_pred)
lr_rmse = lr_mse ** 0.5
```

```python
# 4. KNN Regresyon
knn_model = KNeighborsRegressor(n_neighbors=5)
knn_model.fit(X_train, y_train)
knn_pred = knn_model.predict(X_test)
knn_mse = mean_squared_error(y_test, knn_pred)
knn_rmse = knn_mse ** 0.5
```

```python
# Sonuçları yazdırma
dict_results = {
    'Random Forest (Reduced Trees)': {'MSE': rf_mse_reduced, 'RMSE': rf_rmse_reduced},
    'Random Forest (Small Data)': {'MSE': rf_mse_small, 'RMSE': rf_rmse_small},
    'Linear Regression': {'MSE': lr_mse, 'RMSE': lr_rmse},
    'KNN': {'MSE': knn_mse, 'RMSE': knn_rmse},
}

for model_name, result in dict_results.items():
    print(f"{model_name} - MSE: {result['MSE']}, RMSE: {result['RMSE']}")
```

```python
# Gerçek ve tahmin edilen değerleri almak
rf_pred_reduced = rf_model_reduced.predict(X_test)
rf_pred_small = rf_model_small.predict(X_test)
lr_pred = lr_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
# Hata analizi hesaplamalarını yapalım
def compute_error_metrics(y_test, y_pred):
    mse = mean_squared_error(y_test, y_pred)
    rmse = mse ** 0.5
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    return mse, rmse, mae, r2
```

```python
# Model sonuçları için hata analizi
error_metrics = {}

# Random Forest (Reduced Trees)
mse_rf, rmse_rf, mae_rf, r2_rf = compute_error_metrics(y_test, rf_pred_reduced)
error_metrics['Random Forest (Reduced Trees)'] = {'MSE': mse_rf, 'RMSE': rmse_rf, 'MAE': mae_rf, 'R²': r2_rf}

# Random Forest (Small Data)
mse_rf_small, rmse_rf_small, mae_rf_small, r2_rf_small = compute_error_metrics(y_test, rf_pred_small)
error_metrics['Random Forest (Small Data)'] = {'MSE': mse_rf_small, 'RMSE': rmse_rf_small, 'MAE': mae_rf_small, 'R²': r2_rf_small}

# Linear Regression
mse_lr, rmse_lr, mae_lr, r2_lr = compute_error_metrics(y_test, lr_pred)
error_metrics['Linear Regression'] = {'MSE': mse_lr, 'RMSE': rmse_lr, 'MAE': mae_lr, 'R²': r2_lr}

# KNN
mse_knn, rmse_knn, mae_knn, r2_knn = compute_error_metrics(y_test, knn_pred)
error_metrics['KNN'] = {'MSE': mse_knn, 'RMSE': rmse_knn, 'MAE': mae_knn, 'R²': r2_knn}
```

```python
# Sonuçları yazdırma
for model_name, result in error_metrics.items():
    print(f"{model_name} - MSE: {result['MSE']}, RMSE: {result['RMSE']}, MAE: {result['MAE']}, R²: {result['R²']}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Modellerin hata metriklerini görselleştirmek
models = list(error_metrics.keys())
mse_values = [error_metrics[model]['MSE'] for model in models]
rmse_values = [error_metrics[model]['RMSE'] for model in models]
mae_values = [error_metrics[model]['MAE'] for model in models]
r2_values = [error_metrics[model]['R²'] for model in models]
```

```python
# Grafik düzeni için X ekseninin aralığını ayarlıyoruz
x = np.arange(len(models))

# 4 farklı metrik için tek bir görselleştirme
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# 1. MSE
axs[0, 0].bar(x, mse_values, color='lightblue')
axs[0, 0].set_title('MSE (Mean Squared Error)')
axs[0, 0].set_xticks(x)
axs[0, 0].set_xticklabels(models, rotation=45, ha='right')
for i, v in enumerate(mse_values):
    axs[0, 0].text(x[i], v + 5, f'{v:.2f}', ha='center', va='bottom')

# 2. RMSE
axs[0, 1].bar(x, rmse_values, color='lightgreen')
axs[0, 1].set_title('RMSE (Root Mean Squared Error)')
axs[0, 1].set_xticks(x)
axs[0, 1].set_xticklabels(models, rotation=45, ha='right')
for i, v in enumerate(rmse_values):
    axs[0, 1].text(x[i], v + 0.5, f'{v:.2f}', ha='center', va='bottom')

# 3. MAE
axs[1, 0].bar(x, mae_values, color='lightcoral')
axs[1, 0].set_title('MAE (Mean Absolute Error)')
axs[1, 0].set_xticks(x)
axs[1, 0].set_xticklabels(models, rotation=45, ha='right')
for i, v in enumerate(mae_values):
    axs[1, 0].text(x[i], v + 0.5, f'{v:.2f}', ha='center', va='bottom')

# 4. R²
axs[1, 1].bar(x, r2_values, color='lightyellow')
axs[1, 1].set_title('R² (R-squared)')
axs[1, 1].set_xticks(x)
axs[1, 1].set_xticklabels(models, rotation=45, ha='right')
for i, v in enumerate(r2_values):
    axs[1, 1].text(x[i], v + 0.02, f'{v:.2f}', ha='center', va='bottom')

# Grafik düzenini ayarla
fig.tight_layout()

# Görselleştirmeyi göster
plt.show()
```
